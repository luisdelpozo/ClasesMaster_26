## EXERCICE 2

```python
def count(file, pattern):
    

    f = open(file)
    lines = f.readlines()
    count=0
    for line in lines:
        if line.find(pattern) != -1:
            count += 1
    return count

path = '/home/dsc/Data/shell/unzipped/Finn.txt'
        
```

```python
count(path,'Huck')
```

```python
help(str.find)
```

```python
def count_2(file, pattern):
    
    with open(file) as f:
        lines = f.readlines()
        count=0
        
    for line in lines:
        if pattern in line:
            count += 1
    
    print(f.closed)
    
    return count

path = '/home/dsc/Data/shell/unzipped/Finn.txt'
```

```python
count_2(path,'Huck')
```

```python
def process_a(lines):
    
    result = []
    for line in lines:
        if len(line) > 1:
            result.append(last_word(line))
        
    return result 

process_a(lines)
```

```python
def process_b(lines):
    
    result = []
    for line in lines:
        if len(line) > 1:
            result.append(line[0])
        
    return result 

process_b(lines)
```

```python
def process(f, lines):
    
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(f(line))
        
    return result 

process(len,lines)
```

```python
list_names = ['Luis','Carola','Celia','Sara','PAco','Doris','Fabián','Here','HECTOR','sole']
```

```python
help(list_names[0].lower)
```

```python
#INTENTO FALLIDO

vowels = ['a','e','i','o','u']

for name in list_names:
    i = 0
    for vowel in vowels:
        j = 0
        if name[j] == vowel:
            name[j] = vowel.upper
        else:
            name[j] = vowel.lower
        j =+ 1
    
        list_names[i] = name
    
        i =+ 1

list_names           
```

```python
names = ['Luis','Carola','Celia','Sara','PAco','Doris','Fabián','Here','HECTOR','sole']
```

```python
result = []

for name in names:
    partial_result = ''
    
    for letter in name:
        if letter in 'aeiouAEIOU':
            partial_result += letter.upper()
        else:
            partial_result += letter.lower()
                
    result.append(partial_result)

result
```

# 3

```python
def my_filter(character, collection):
    result = []
    for element in collection:
        if character.lower() in element.lower():
            result.append(element)
    return result

my_filter('c', names)
```

# LIST COMPRENHENSIONS

```python
[ n ** 2 for n in numbers]
```

```python
result = []

for n in numbers:
    result.append(n**2)
result
```

```python
list(map(lambda n:n**2,numbers))
```

```python
# DICT COMPRENHENSION

{n: n ** 2 for n in numbers}
```

Con filtro

```python
[ n ** 2 for n in numbers if n % 2==0]
```

```python
result = []

for n in numbers:
    if n % 2==0:
        result.append(n**2)
result
```

```python
# ESto es un cagarro seco en comparación con lo anterior

list(map(lambda n:n**2,filter(lambda n:n % 2 == 0,numbers)))
```
