```python
def process_a(lines):
    
    result = []
    for line in lines:
        if len(line) > 1:
            result.append(last_word(line))
        
    return result 

process_a(lines)
```

```python
def process_b(lines):
    
    result = []
    for line in lines:
        if len(line) > 1:
            result.append(line[0])
        
    return result 

process_b(lines)
```

```python
def process(f, lines):
    
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(f(line))
        
    return result 

process(len,lines)
```

# LIST COMPRENHENSIONS

```python
numbers = list(range(1,10))
```

```python
[ n ** 2 for n in numbers]
```

```python
#ES SIMPLEMENTE UNA MANERA DE AHORRAR ESCRITURA DE CÓDIGO
result = []

for n in numbers:
    result.append(n**2)
result
```

```python
list(map(lambda n:n**2,numbers))
```

```python
# EXISTEN LAS DICT COMPRENHENSION

{n: n ** 2 for n in numbers}
```

### EJEMPLO DE LIST COMPRENHENSION CON FILTRO

```python
[ n ** 2 for n in numbers if n % 2==0]
```

```python
result = []

for n in numbers:
    if n % 2==0:
        result.append(n**2)
result
```

```python
# Esto es un cagarro seco en comparación con lo anterior

list(map(lambda n:n**2,filter(lambda n:n % 2 == 0,numbers)))
```
